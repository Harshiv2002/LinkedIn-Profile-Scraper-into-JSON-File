```python
# LOGIN INTO ACCOUNT
# To start scraping we have to first login into account

from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By

# Creating a webdriver instance
driver = webdriver.Chrome("D:\Software and Setup\chromedriver.exe") #Chrome driver address in pc
# This instance will be used to log into LinkedIn
  
# Opening linkedIn's login page
driver.get("https://linkedin.com/uas/login")

# waiting for the page to load
time.sleep(5)

# entering username
username = driver.find_element(By.ID,"username")

# In case of an error, try changing the element
# tag used here.

# Enter Your Email Address
username.send_keys("<your email goes here>")  

# entering password
pword = driver.find_element(By.ID,"password")
# In case of an error, try changing the element
# tag used here.

# Enter Your Password
pword.send_keys("<your password goes here>")  

# Clicking on the log in button
# Format (syntax) of writing XPath -->
# //tagname[@attribute='value']
driver.find_element(By.XPATH,"//button[@type='submit']").click()
# In case of an error, try changing the
# XPath used here.
```

```python
# FUNCTION TO SCROLL AND LOAD WEBPAGE
# This scroll function is used to completely load the webpage before scraping

def scroll():
    start = time.time()

    # will be used in the while loop
    initialScroll = 0
    finalScroll = 1000

    while True:
    	driver.execute_script(f"window.scrollTo({initialScroll},{finalScroll})")
    	# this command scrolls the window starting from
    	# the pixel value stored in the initialScroll
    	# variable to the pixel value stored at the
    	# finalScroll variable
    	initialScroll = finalScroll
    	finalScroll += 1000

    	# we will stop the script for 1.5 seconds so that
    	# the data can load
    	time.sleep(1.5)
    	# You can change it as per your needs and internet speed  

    	end = time.time()

    	# We will scroll for 20 seconds.
    	# You can change it as per your needs and internet speed
    	if round(end - start) > 10:
    		break
```

```python
# FUNCTION TO SCRAPE FROM LINKEDIN
# To keep things simple we will scrape name and job description from the LinkedIn account

def scrape(job_des):
    src = driver.page_source
    soup = BeautifulSoup(src, 'lxml')
    
    intro = soup.find('div', {'class': 'pv-text-details__left-panel'})
    about = soup.find('div', {'class': 'display-flex ph5 pv3'})
    
    # In case of an error, try changing the tags used here.

    name_loc = intro.find("h1")

    # Extracting the Name
    name = name_loc.get_text().strip()
    # strip() is used to remove any extra blank spaces

    works_at_loc = intro.find("div", {'class': 'text-body-medium'})

    # this gives us the HTML of the tag in which the Company Name is present
    # Extracting the Company Name
    works_at = works_at_loc.get_text().strip()


    des_loc = about.find_all("span", {'aria-hidden': 'true'})
    des = []

    for i in des_loc:
        des.append(i.get_text())
    
    # filter below
    # name and description will be automatically scrapped and job title will be taken from dictonary 
    Dict = dict({"name": name,"job_title":job_des, "description": works_at})
    #Dict = dict({"name": name, "description": works_at})
    return Dict  
```
